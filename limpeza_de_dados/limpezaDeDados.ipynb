```python
import pandas as pd
import seaborn as srn
import statistics as stc
import numpy as np
```

```python
df = pd.read_csv("tempo.csv", sep=";")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# Iremos excluir linhas vazias, no caso, a linha 9 e a 3

df = df.dropna()
```

```python
# Agora, as colunas colunas Vento e Jogar não precisam mais de alteração, porém
# as outras sim, portanto, iremos aplicar moda em Aperência e mediana na Temp e
# Umidade. O cliente pediu que temp fosse de -130 até 130, enquanto umidade de
# 0 a 100, já a aparência deveria ser sol, nublado ou chuva.
```

```python
moda_aparencia = stc.mode(df["Aparencia"])
moda_aparencia
```

```python
df.groupby(df["Aparencia"]).size()
```

```python
df.loc[df["Aparencia"].isin(["menos"]), "Aparencia"] = moda_aparencia
```

```python
df
```

```python
mediana_temp = stc.median(df["Temperatura"])
df.loc[df["Temperatura"] > 130 , "Temperatura"] = mediana_temp
df.loc[df["Temperatura"] < -130 , "Temperatura"] = mediana_temp
```

```python
df
```

```python
mediana_umi = stc.median(df["Umidade"])
df.loc[df["Umidade"] > 100 , "Umidade"] = mediana_umi
df.loc[df["Umidade"] < 0 , "TUmidade"] = mediana_umi
```
